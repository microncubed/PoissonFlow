# thermal_simulator_slh_var

```python
# Find out current working directory and label sub-directories
import os
cwd = os.getcwd()
print('cwd: ', cwd)

fig_dir = os.path.dirname(cwd) + '/figures/'
data_dir = os.path.dirname(cwd) + '/data/'
config_dir = os.path.dirname(cwd) + '/config/'
src_dir = os.path.dirname(os.path.dirname(os.path.dirname(cwd))) + '/src/'
models_dir = os.path.dirname(os.path.dirname(os.path.dirname(cwd))) + '/models/'
```

```python
os.chdir(src_dir)
from slh_var import model
from pdeSim import PoissonFlow2D
os.chdir(cwd)

import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 16
plt.rcParams['lines.linewidth'] = 3
from matplotlib.colors import LogNorm

import numpy as np
import h5py

# user defined parameters
params = {
    'freq':0,
    'neu_top':True,
    'wet':True,
    'vias':True,
    'power':220,
    'v_max':0.000,
    'heater_metal':True
         }

save_file = True # save?

save_name = 'single_heater_freq_' + str(params['freq']) + '_neu_top_' + str(params['neu_top']) +\
'_wet_' + str(params['wet']) +'_vias_'+str(params['vias']) + '_power_' + str(params['power']) + '_v_max_'+str(params['v_max']) +\
'_heater_metal_' + str(params['heater_metal']) + '_30um_'
print('save name: ',save_name)


# load the model
k_in,c_in,rho_in,Q_in,v_in,x,y,dx,dy,nx,ny,n_bound = model(wet = params['wet'], vias = params['vias'], power = params['power'], 
                                                   heater_metal = params['heater_metal'], v_max = params['v_max'])


print(dy)
```

## Plot heat sources/sinks

```python
# when plotting a pcolor as shading = 'flat' you need these arrays that specify the quad corners
y_flat = np.array([sum(dy[:i]) for i in range(len(dy)+1)])
x_flat = np.array([i*dx for i in range(nx+1)])

fig,ax=plt.subplots(1,1,figsize=[6,4])

myplot = ax.pcolor(x_flat*1e6,y_flat*1e6,Q_in.T,shading='flat')
#ax.set_xlim(x.min()*1e6,x.max()*1e6)
#ax.set_ylim(y.min()*1e6,y.max()*1e6)
#ax.set_ylim(-40,-30)
ax.set_xlabel('x ($\mu$m)')
ax.set_ylabel('y ($\mu$m)')
cbar = fig.colorbar(myplot)
cbar.set_label('q (W/m)')
fig.tight_layout()
#ax.set_xlim([300,700]) 
ax.set_ylim([3550,3750]) 

if save_file==True:
    fig.savefig(fig_dir + save_name+'_q.png',dpi=150,facecolor = 'w')
```

```python
fig,ax=plt.subplots(1,1,figsize=[6,4])
myplot = ax.pcolor(x_flat[::1]*1e6,y_flat[::1]*1e6,k_in.T[::1,::1],shading='flat',norm=LogNorm(vmin=k_in.min(),vmax=k_in.max()))
#ax.set_xlim(x.min()*1e6,x.max()*1e6)
#ax.set_ylim(y.min()*1e6,y.max()*1e6)
#ax.set_ylim(-40,-30)
ax.set_xlabel('x ($\mu$m)')
ax.set_ylabel('y ($\mu$m)')
cbar = fig.colorbar(myplot)
cbar.set_label('$\kappa$ (W/m K)')

ax.set_ylim([3550,3750]) 
#ax.set_xlim([300,700]) 
fig.tight_layout()
if save_file==True:
    fig.savefig(fig_dir + save_name+'_zoom_-5_kappa.png',dpi=150,facecolor = 'w')
    
```

## Figure out thermal resistance

```python
depth = 1e-4


Q_input = np.sum(Q_in)*dx*1.5e-7 * 1e-4
print(Q_input)

dT = np.average(out.real.T[24,90:110])
R_th = dT/Q_input/1000
print('Simulated thermal resistance {:0.3f} (K/mW)'.format(R_th))
```

## Save, if needed

```python
if save_file==True:
    f=h5py.File(data_dir + save_name+'.hdf5','w')
    
    f.attrs['freq'] = params['freq']
    f.attrs['v_max'] = params['v_max']
    f.attrs['wet'] = params['wet']
    f.attrs['neu_top'] = params['neu_top']
    
    f.attrs['simulated thermal resistance'] = R_th
    
    f.attrs['dx'] = dx
    #f.attrs['dy'] = dy
    f.attrs['nx'] = nx
    f.attrs['ny'] = ny
    
    #f.attrs['notes']= np.string_('some comments')
    
    dset_dy = f.create_dataset("dy", (ny,), dtype='f')
    dset_dy[...]=dy
    
    dset_v = f.create_dataset("v", (nx,ny), dtype='f')
    dset_v[...]=v_in

    dset_T = f.create_dataset("T", (nx,ny), dtype='f')
    dset_T[...]=out.real

    dset_kappa = f.create_dataset("kappa", (nx,ny), dtype='f')
    dset_kappa[...]=k_in
    
    dset_Q_in = f.create_dataset("Q_in", (nx,ny), dtype='f')
    dset_Q_in[...]=Q_in

    dset_c_in = f.create_dataset("c", (nx,ny), dtype='f')
    dset_c_in[...]=c_in
    
    dset_rho_in = f.create_dataset("rho", (nx,ny), dtype='f')
    dset_rho_in[...]=rho_in

    f.close()
```
